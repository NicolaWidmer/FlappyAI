```python
layer_sizes = [10,20,30,70,80,150]
rounds = 500
for ls in layer_sizes:
    rewards = []
    meanrewards = []
    agent = DeepQ.ClippedDoubleDQN.ClippedDoubleDQN(ls,0.001,0.99,0.05,0.01)
    
    print(f"layer_size {ls}")
    for i in range(rounds):
        if i%10 == 0:
            print(i)
        if (i+1)%100 == 0:
            #game.make_video(lambda x: agent3.get_action(x,train=False),f"3_episode{i+1}")
            print(i)
            print(sum(rewards[-100:])/100)
        rewards.append(main.run_episode(game,agent,True,1000))
        meanrewards.append(mean(rewards[-100:]))
    plt.plot(meanrewards,label=f"{rounds} rounds, layer_size {ls}")
plt.legend()
```

```python
taus = [0.001,0.01,0.1,0.9,1]
rounds = 600
for tau in taus:
    rewards = []
    meanrewards = []
    agent = DeepQ.ClippedDoubleDQN.ClippedDoubleDQN(layer_size,0.001,0.99,0.05,tau)
    
    print(f"tau: {tau}")
    for i in range(rounds):
        if i%10 == 0:
            print(i)
        if (i+1)%100 == 0:
            #game.make_video(lambda x: agent3.get_action(x,train=False),f"3_episode{i+1}")
            print(i)
            print(sum(rewards[-100:])/100)
        rewards.append(main.run_episode(game,agent,True,1000))
        meanrewards.append(mean(rewards[-100:]))
    plt.plot(meanrewards,label=f"{rounds} rounds, tau: {tau}")
plt.legend()
```
